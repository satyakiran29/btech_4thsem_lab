```python
import pandas as pd
import math
data=pd.read_csv('p.csv')
print("p.satya kiran ")
print("24345A0524 CSE-C")
def mean(X):
  sum=0
  for i in range(len(X)):
    sum=sum+X[i]
  return sum/len(X) 
def variance(X,avg):
  var_sum=0
  for i in range(len(X)):
    var_sum=var_sum+((X[i]-avg)**2)
  return var_sum/len(X) 
def minimum(X):
  m=X[0]
  for i in range(len(X)):
    if (m>X[i]):
      m=X[i]
  return m    
def maximum(X):
  m=X[0]
  for i in range(len(X)):
    if (m<X[i]):
      m=X[i]
  return m    
def median(X):
  n=len(X)
  if n % 2 == 0: 
    m1=sorted(X)[n//2] 
    m2=sorted(X)[n//2 - 1] 
    return (m1+m2)/2
  else:
    return sorted(X)[n//2]
def percentile_25(X):
  n=len(X)
  if n % 4 == 0: 
    m1=sorted(X)[n//4] 
    m2=sorted(X)[n//4 - 1] 
    return (m1+m2)/2
  else:
    return sorted(X)[n//4]
def percentile_75(X):
  n=len(X)
  if (3*n) % 4 == 0: 
    m1=sorted(X)[(3*n)//4] 
    m2=sorted(X)[(3*n)//4 - 1] 
    return (m1+m2)/2
  else:
    return sorted(X)[(3*n)//4]    
attributes=data.columns.tolist()
for i in range(len(attributes)-1):
  s=attributes[i]
  print("\n",attributes[i])
  X=data[s]
  avg=mean(X)
  print("MEAN:",avg)
  var=variance(X,avg)
  print("VARIANCE:",var)
  std=math.sqrt(var)
  print("STANDARD DEVIATION:",std)
  min=minimum(X)
  print("MINIMUM VALUE:",min)
  q1=percentile_25(X)
  print("25th PERCENTILE:",q1)
  m=median(X)
  print("50th PERCENTILE(median):",m)
  q2=percentile_75(X)
  print("75th PERCENTILE:",q2)
  max=maximum(X)
  print("MAXIMUM VALUE:",max)
```
