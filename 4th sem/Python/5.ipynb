```python
import numpy as np
import matplotlib.pyplot as plt
n_rolls=1000
rolls=np.random.randint(1,7,n_rolls)
theoretical_probabilities=np.ones(6)/6
empirical_probabilities=[np.sum(rolls==i)/n_rolls for i in range(1,7)]
x=np.arange(1,7)
plt.bar(x-0.2,theoretical_probabilities,width=0.4,label='Theoretical',color='blue',alpha=0.5)
plt.bar(x+0.2,empirical_probabilities,width=0.4,label='Empirical',color='red',alpha=0.5)
plt.xlabel('Die Face')
plt.ylabel('Probability')
plt.title('Theoretical vs Empirical Probabilities of Die Rolls')
plt.xticks(x)
plt.legend()
plt.show()
```
