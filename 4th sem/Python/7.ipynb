```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import bernoulli
print("p.satya kiran ")
print("24345A0524 CSE-C")

data = bernoulli.rvs(p=0.2, size=1000)


fig, ax = plt.subplots()
ax.hist(data, bins=2, alpha=0.5, color='blue', edgecolor='black')
ax.set(xlabel='Bernoulli Outcome', ylabel='Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
data = np.random.binomial(10,0.5,50)
print("p.satya kiran ")
print("24345A0524 CSE-C")
plt.hist(data, bins=10, alpha=0.5, color='blue', edgecolor='black')
plt.xlabel('Binomial Outcome')
plt.ylabel('Frequency')
plt.title('Histogram of Binomial Distribution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.poisson(3, 50)
plt.hist(data, bins=10, alpha=0.5, color='blue', edgecolor='black')
print("p.satya kiran ")
print("24345A0524 CSE-C")
plt.xlabel('Poisson Outcome')
plt.ylabel('Frequency')
plt.title('Histogram of Poisson Distribution')
plt.show()
```
